# ***9. Exploring Data based on Certain Criteria and Graph Plotting***

Now we go deep into into the analysis of our data. the analysis involving sorting and grouping of beer basid on certain questions that are mentioned below.

```python
df1.describe()
```

I am planning to conduct data exploration to analyze which beer sells the most and identify the locations where these sales are highest.

a) finding the beers sale in bottels

```python
a=df1[df1['SKU']=="Bottles"]
data1 = a[(a['Bitterness'] > 55)&(a['Fermentation_Time']>18)&(a['pH_Level']>5)&(a['Volume_Produced']>3500)]
data1
```

```python
sit = data1[['Brew_Date', 'Beer_Style', 'Total_Sales']]
kf = pd.DataFrame(sit)
sales_by_beer_style = kf.groupby('Beer_Style')['Total_Sales'].sum()
total_styles = len(sales_by_beer_style)
myexplode = [0.1] * total_styles
plt.figure(figsize=(8, 6))
plt.pie(sales_by_beer_style, labels=sales_by_beer_style.index, explode=myexplode, shadow=False, autopct='%1.1f%%', startangle=100)
plt.title('Distribution of Total Sales by Beer Style')
plt.legend(title = "Beer:",loc='upper left')
plt.axis('equal')
plt.show()

```

Based on the data exploration, it can be inferred that IPA beer in bottles from Rajajinaga has higher sales compared to other beer varieties

```python
sit = data1[['Brew_Date', 'Beer_Style', 'Total_Sales']]
cf = pd.DataFrame(sit)
sss = cf.groupby('Beer_Style')['Total_Sales'].sum().reset_index()
plt.figure(figsize=(10, 6))
sns.countplot(data=cf, x='Beer_Style', palette='Set1',hue='Beer_Style',legend=False)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


```

b) Taking beer in Kegs

```python
k=df1[df1['SKU']=="Kegs"]
data2= k[(k['Bitterness'] > 55)&(k['Fermentation_Time']>18)&(k['pH_Level']>5)&(k['Volume_Produced']>3500)]
data2
```

```python
sip = data2[['Brew_Date', 'Total_Sales', 'Beer_Style']]
lf = pd.DataFrame(sip)
lf['Brew_Date'] = pd.to_datetime(df['Brew_Date'])
sales_by_beer = lf.groupby('Beer_Style')['Total_Sales'].sum()
plt.figure(figsize=(12, 8))
sns.boxplot(data=sip, x='Beer_Style', y='Total_Sales', palette='Set1',hue='Beer_Style',legend=False)
plt.title('Total Sales of Beer Styles')
plt.xlabel('Beer Style')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

beer_with_highest_sales = sales_by_beer.idxmax()
highest_sales_value = sales_by_beer.max()
print(f"The beer with the highest total sales is '{beer_with_highest_sales}' with total sales of {highest_sales_value:.2f}.")
```

Based on the data exploration, it can be inferred that Pilsner beer in kegs from Electonic City has higher sales compared to other beer varieties

c) Taking Cans

```python
c=df1[df1['SKU']=="Cans"]
data3= c[(c['Bitterness'] > 55)&(c['Fermentation_Time']>18)&(c['pH_Level']>5)&(c['Volume_Produced']>3500)]
# data3
```

```python
sub = data3[['Brew_Date', 'Total_Sales','Beer_Style']]
pf= pd.DataFrame(sub)
pf['Brew_Date'] = pd.to_datetime(df['Brew_Date'])
sale = pf.groupby('Beer_Style')['Total_Sales'].sum().reset_index()
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x='Beer_Style', y='Total_Sales', data=sale, estimator=sum, errorbar=None)
plt.title('Total Sales by Beer Style')
plt.xlabel('Beer Style')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
all_data=sale
# sales_by_beer = pf.groupby('Beer_Style')['Total_Sales'].sum()
beer_with_highest_sales = all_data['Total_Sales'].idxmax()
highest_sales_value = all_data['Total_Sales'].max()
print(f"The beer with the highest total sales is '{all_data.loc[beer_with_highest_sales, 'Beer_Style']}' with total sales of {highest_sales_value:.2f}.")
```

Based on the data exploration, it can be inferred that Ale beer in Cans has higher sales compared to other beer varieties

```python
p=df1[df1['SKU']=="Pints"]
data4= p[(p['Bitterness'] > 55)&(p['Fermentation_Time']>18)&(p['pH_Level']>5)&(p['Volume_Produced']>3500)]
data4
```

```python
subb=data4[['Brew_Date', 'Total_Sales','Beer_Style']]
tf = pd.DataFrame(subb)
tf['Brew_Date'] = pd.to_datetime(df['Brew_Date'])
sales_by_beer = tf.groupby('Beer_Style')['Total_Sales'].sum().reset_index()
sns.set_style("whitegrid")
plt.figure(figsize=(10, 6))
sns.pointplot(x='Beer_Style', y='Total_Sales', data=sales_by_beer, color='b')
plt.title('Total Sales by Beer Style')
plt.xlabel('Beer Style')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

sales_by_beer = tf.groupby('Beer_Style')['Total_Sales'].sum()
beer_with_highest_sales = sales_by_beer.idxmax()
highest_sales_value = sales_by_beer.max()
print(f"The beer with the highest total sales is '{beer_with_highest_sales}' with total sales of {highest_sales_value:.2f}.")
```

Based on the data exploration, it can be inferred that Ale beer in Pints has higher sales compared to other beer varieties

```python
dfs = [data1, data2, data3, data4]
all_data = pd.concat(dfs)
total = all_data.groupby(['Beer_Style', 'Location','SKU'])['Total_Sales'].sum().reset_index()
max= total.loc[total['Total_Sales'].idxmax()]
```
